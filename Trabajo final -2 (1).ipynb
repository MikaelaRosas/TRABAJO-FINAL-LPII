```python
#Todos los anuncios de las ofertas en una lista
Ofertas= driver.find_elements_by_xpath('//li[@data-offer-type="Cache"]')
```

```python
for oferta in Ofertas:
        precio= oferta.find_element_by_xpath('.//span[@class="price-amount"]').text
        #if precio != oferta.find_element_by_xpath('.//span[@class="price-amount"]' or class="count").text:
            #precio= NULL
        #else:
        #print(precio)

        Partida= oferta.find_element_by_xpath('.//div[@class="arrival-country"]').text
        #if Partida!= oferta.find_element_by_xpath('.//div[@class="departure"]').text:
            #Partida= NULL
        #else:
        #print(Partida)

        Destino= oferta.find_element_by_xpath('.//div[@class="arrival-city"]').text
            #if Destino!= oferta.find_element_by_xpath('.//div[@class="arrival-city"]').text:
            #Destino= NULL
        #else:
        print("-------------------------------------")
        print("[Partida]:", Partida)
        print("[destino]:",Destino)
        print("[Precio]:",precio)
```

```python
Origen1=[]
for oferta in Ofertas:
    Origen= oferta.find_element_by_xpath('.//div[@class="title"]').text
    if Origen !="":
        Origen1.append(Origen)

Origen2=[]
for i in Origen1:
    if i != "desde: Lima (LIM)":
        i = "desde: Lima (LIM)"
    Origen2.append(i)
#    print(i)
#print(Origen2)

Origen=[]
for j in Origen2:
#    print(j)
    if j != '':
        j = "('PERÚ', 'Lima')"
    Origen.append(j)

for i in Origen:
    print(i)
```

```python
Departamento=[]
for oferta in Ofertas:
    Destino= oferta.find_element_by_xpath('.//div[@class="arrival-city"]').text
    if Destino !="":    
        Departamento.append(Destino)
for i in Departamento:
    print(i)
```

```python
Pais1=[]
for oferta in Ofertas:
    Pais= oferta.find_element_by_xpath('.//div[@class="arrival-country"]').text
    if Pais != "":
        Pais1.append(Pais)

for j in Pais1:
    print(j)
```

```python
n=len(Departamento)
Destino=[]
for i in range(0,n):
    j=Pais1[i],Departamento[i]
    Destino.append(j)

for i in Destino:
     print(i)
```

```python
Precio=[]
for oferta in Ofertas:
    precio= oferta.find_element_by_xpath('.//span[@class="price-amount"]').text
    if precio != "":
        Precio.append(precio)

Costo=[]
for j in Precio:
    if j != float or int:
        j = j+" dolares"
    Costo.append(j)

for i in Costo:
    print(i)
```
